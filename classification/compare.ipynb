```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
log_name1 = os.path.join("./checkpoints/pix2pix/resnet18_True", "log.txt")
log_name2 = os.path.join("./checkpoints/cyclegan/resnet18_True", "log.txt")
log_name3 = os.path.join("./checkpoints/pix2pixHD/resnet18_True", "log.txt")
log_name4 = os.path.join("./checkpoints/pix2pix_dct/resnet18_True", "log.txt")
log_name5 = os.path.join("./checkpoints/aggan/resnet18_True", "log.txt")
```

```python
import re

tmp = {}
log1 = []
log2 = []
log3 = []
log4 = []
log5 = []


f = open(log_name1, 'r')
sourceInlines = f.readlines()
for i, line in enumerate(sourceInlines):
    
    if i % 3 == 0:
        tmp = {}
        tmp['epoch'] = int(re.split('[ -]', line)[1])
    elif i % 3 == 1:
        split_list = line.strip('\n').split(' ')
        # print(split_list)
        tmp['train_loss'] = float(split_list[2])
        tmp['train_acc'] = float(split_list[4])
        # print(tmp['train_loss'])
    elif i % 3 == 2:
        split_list = line.strip('\n').split(' ')
        # print(split_list)
        tmp['val_loss'] = float(split_list[2])
        tmp['val_acc'] = float(split_list[4])
    
        log1.append(tmp)
    
f = open(log_name2, 'r')
sourceInlines = f.readlines()
for i, line in enumerate(sourceInlines):
    
    if i % 3 == 0:
        tmp = {}
        tmp['epoch'] = int(re.split('[ -]', line)[1])
    elif i % 3 == 1:
        split_list = line.strip('\n').split(' ')
        # print(split_list)
        tmp['train_loss'] = float(split_list[2])
        tmp['train_acc'] = float(split_list[4])
        # print(tmp['train_loss'])
    elif i % 3 == 2:
        split_list = line.strip('\n').split(' ')
        # print(split_list)
        tmp['val_loss'] = float(split_list[2])
        tmp['val_acc'] = float(split_list[4])
    
        log2.append(tmp)

f = open(log_name3, 'r')
sourceInlines = f.readlines()
for i, line in enumerate(sourceInlines):
    
    if i % 3 == 0:
        tmp = {}
        tmp['epoch'] = int(re.split('[ -]', line)[1])
    elif i % 3 == 1:
        split_list = line.strip('\n').split(' ')
        # print(split_list)
        tmp['train_loss'] = float(split_list[2])
        tmp['train_acc'] = float(split_list[4])
        # print(tmp['train_loss'])
    elif i % 3 == 2:
        split_list = line.strip('\n').split(' ')
        # print(split_list)
        tmp['val_loss'] = float(split_list[2])
        tmp['val_acc'] = float(split_list[4])
    
        log3.append(tmp)
    
f = open(log_name4, 'r')
sourceInlines = f.readlines()
for i, line in enumerate(sourceInlines):
    
    if i % 3 == 0:
        tmp = {}
        tmp['epoch'] = int(re.split('[ -]', line)[1])
    elif i % 3 == 1:
        split_list = line.strip('\n').split(' ')
        # print(split_list)
        tmp['train_loss'] = float(split_list[2])
        tmp['train_acc'] = float(split_list[4])
        # print(tmp['train_loss'])
    elif i % 3 == 2:
        split_list = line.strip('\n').split(' ')
#         print(split_list)
#         print(val_acc1.index(max(val_acc1)))
#         print(val_acc2.index(max(val_acc2)))
#         print(val_acc3.index(max(val_acc3)))
#         print(val_acc4.index(max(val_acc4)))
#         print(val_acc5.index(max(val_acc5)))
#         print(val_acc6.index(max(val_acc6)))
        tmp['val_loss'] = float(split_list[2])
        tmp['val_acc'] = float(split_list[4])
    
        log4.append(tmp)

f = open(log_name5, 'r')
sourceInlines = f.readlines()
for i, line in enumerate(sourceInlines):
    if i % 3 == 0:
        tmp = {}
        tmp['epoch'] = int(re.split('[ -]', line)[1])
    elif i % 3 == 1:
        split_list = line.strip('\n').split(' ')
        # print(split_list)
        tmp['train_loss'] = float(split_list[2])
        tmp['train_acc'] = float(split_list[4])
        # print(tmp['train_loss'])
    elif i % 3 == 2:
        split_list = line.strip('\n').split(' ')
        # print(split_list)
        tmp['val_loss'] = float(split_list[2])
        tmp['val_acc'] = float(split_list[4])
    
        log5.append(tmp)
    
```

```python
epoch = [x['epoch'] for x in log1]

train_loss1 = [x['train_loss'] for x in log1]
train_acc1 = [x['train_acc'] for x in log1]
val_loss1 = [x['val_loss'] for x in log1]
val_acc1 = [x['val_acc'] for x in log1]

train_loss2 = [x['train_loss'] for x in log2]
train_acc2 = [x['train_acc'] for x in log2]
val_loss2 = [x['val_loss'] for x in log2]
val_acc2 = [x['val_acc'] for x in log2]

train_loss3 = [x['train_loss'] for x in log3]
train_acc3 = [x['train_acc'] for x in log3]
val_loss3 = [x['val_loss'] for x in log3]
val_acc3 = [x['val_acc'] for x in log3]

train_loss4 = [x['train_loss'] for x in log4]
train_acc4 = [x['train_acc'] for x in log4]
val_loss4 = [x['val_loss'] for x in log4]
val_acc4 = [x['val_acc'] for x in log4]

train_loss5 = [x['train_loss'] for x in log5]
train_acc5 = [x['train_acc'] for x in log5]
val_loss5 = [x['val_loss'] for x in log5]
val_acc5 = [x['val_acc'] for x in log5]
```

```python
# test accuracy
%matplotlib notebook
plt.plot(epoch[:50:2], val_acc1[:50:2], '--ko', label = 'Pix2pix',linewidth=1.3, markersize = 5)
plt.plot(epoch[:50:2], val_acc2[:50:2], '--yd', label = 'CycleGAN',linewidth=1.3, markersize = 5)
plt.plot(epoch[:50:2], val_acc3[:50:2], '--bs', label = 'Pix2pixHD',linewidth=1.3, markersize = 5)
plt.plot(epoch[:50:2], val_acc4[:50:2], '--c*', label = 'FGGAN',linewidth=1.3, markersize = 5)
plt.plot(epoch[:50:2], val_acc5[:50:2], '--r^', label = 'AGGAN',linewidth=1.3, markersize = 5)
plt.xlim(0,50)
plt.ylim(0.5,1)
plt.xlabel('epoch',fontdict={'weight':'normal','size': 14})
plt.ylabel('accuracy',fontdict={'weight':'normal','size': 14})
plt.tick_params(labelsize=14)
plt.grid()
plt.legend()
plt.savefig('./Figure/test_acc.png')
```
